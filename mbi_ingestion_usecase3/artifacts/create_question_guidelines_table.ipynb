```python
from pyspark.sql.functions import col, expr, monotonically_increasing_id, lit, uuid
from pyspark.sql.types import StructType, StructField, StringType, ArrayType


df_raw = (spark.read
    .option("multiLine", "true")
    .json("/Volumes/serverless_stable_14ey07_catalog/mbi/vol1/source/usecase3/supporting_files/Q&G_uc3.json")
)

display(df_raw)
```

```python

# ── 2. Transform to match target schema ───────────────────────────────────────
df_transformed = df_raw \
    .select(
        expr("uuid()").alias("eval_id"),          # unique UUID per row
        col("question").alias("request"),
        col("guidelines"),
        lit(None).cast(StringType()).alias("metadata"),
        lit(None).cast(StringType()).alias("tags")
    )

display(df_transformed)
```

```python
df_transformed.write.format("delta").mode("overwrite").saveAsTable("serverless_stable_14ey07_catalog.mbi.usecase3_question_guidelines")
```
